```python
def average_by_key(rows):
    total = 0
    count = 0                

    for r in rows:
        if "score" in r:      
            total += r["score"]
            count += 1       

    if count == 0:            
        return 0

    return total / count


print(average_by_key([{"score": 5}, {"score": 7}, {"x": 1}]))  # Expected Output: 6.0
print(average_by_key([]))   # Expected Output: 0
```

```python
def even_numbers(nums):
    even = []
    for n in nums:
        if n % 2 == 0:
            even.append(n)
    return even     

print(even_numbers([1,2,3,4,5,6])) # Expected Output: [2, 4, 6]
```

```python
class Artist:
    def __init__(self, name, fee):
        self.name = name
        self.__fee = fee #private attribute, do not change this.

    def add_charges(self, additional_charges):
        self.__fee = self.__fee + additional_charges
    
    def get_fees(self):
        return self.__fee

a1 = Artist("Taylor Swift", 50)
a2 = Artist("Lisa", 65)
a1.add_charges(20)
a2.add_charges(20)
print("Taylor Swift fee=", a1.get_fees()) # Expected Output: 70
print("Lisa fee=", a2.get_fees()) # Expected Output: 85
```

Develop a python program to create text file with three lines. (File name - Student_ID_CVE.txt)

Hello! Welcome to KMUTT.

We are Civil Engineering Students.

We don't want to learn programming T_T

```python
# Write your program here
class open_read():
    def __init__(self, path):
        self.path = path
        self.f = None
    def __enter__(self):
        self.f = open(self.path, 'r')
        return self.f
    def __exit__(self, exc_type, exc, tb):
        if self.f and not self.f.closed:
            self.f.close()
        return True

text_path = "Student_ID_CVE.txt" 
with open(text_path, "w", encoding="utf-8") as w:
    w.write("Hello! Welcome to KMUTT\nWe are Civil Engineering Students\nWe don't want to learn programming T_T")
```
