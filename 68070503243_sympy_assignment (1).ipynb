```python
#write your code
# --- Part 1: Basic Algebra and Substitution ---

# 1. Define symbolic variable and expression

import numpy as np
import sympy as sp

x= sp.Symbol('x')
E_x= 3*x**2+5*x-7
E_x
```

```python
#write your code
# --- Part 2: Solving Equations ---
# 1. Solve a Quadratic Equation: x^2 - 4x + 3 = 0


x= sp.symbols("x")
Quad_Eq = x**2-4*x+3
solution_Qu= sp.solve(Quad_Eq,x)
solution_Qu
```

Tasks:

Find the Derivative: Define a function $g(x) = 4x^3 - 6x^2$. Use diff() to find the first derivative, $g'(x)$.

Find the Integral: Use integrate() to find the indefinite integral of the original function $g(x) = 4x^3 - 6x^2$. (Note: SymPy does not automatically add the constant of integration $+C$.)

```python
#write your code
# --- Part 3: Fundamental Calculus ---
# 1. Solve function g(x)
G = 4*x**3 - 6*x**2
G_answer = sp.solve(4*x**3 - 6*x**2)
G_answer
```

```python
# 1. Find the Derivative
x = sp.Symbol("x")
derivative = sp.diff( G, x )
derivative
```

```python
# 2. Find the Indefinite Integral
integrate = sp.integrate(G, x)
integrate
```
